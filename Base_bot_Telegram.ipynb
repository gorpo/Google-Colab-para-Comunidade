```
import telepot
import subprocess
import time

#api do bot
api = ''
bot = telepot.Bot(api)


# função do bot
def funcaoBot(msg):
    content_type, chat_type, chat_id = telepot.glance(msg)
    if msg.get('text'):
        texto = msg['text']
        #cada if é uma mensagem que o bot vai responder      
        if 'oi' in texto:                     
            bot.sendMessage(chat_id, 'Olá')
        if 'corno' in texto:                     
            bot.sendMessage(chat_id, 'Você otário')
        if 'python' in texto:                     
            bot.sendMessage(chat_id, 'É como fui programado')

#chama a função e deixa o bot em um loop para ficar ativo   
bot.message_loop(funcaoBot)
print ('Bot esta online, para desativar o bot feche o programa!')
while 1:
        pass




```
