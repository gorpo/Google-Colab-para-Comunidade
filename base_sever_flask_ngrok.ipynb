Codigo base do Flask

```
from pyngrok import ngrok
from flask import Flask

#inicia uma instancia do flask
app = Flask(__name__)

#cria um link do ngrok na porta 5000
link_ngrok = ngrok.connect(5000)
print(f'Link compartilhavel: {link_ngrok}')


#cria uma pagina simples em html
@app.route("/")
def home():
    return "<h1>Meu primeiro site em Flask no Google Colab!</h1>"
  
app.run()
```
